```python

import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.figure(figsize=(1, 1))
```

```python
def gaussian(x,miu,std):
    return (1/(std*((2*np.pi)**(1/2))))*np.exp(-0.5*((x-miu)/std)**(1/2))
```

```python
def my_gaussian(x,miu,std):
    return (1/(std*np.sqrt(2*np.pi)))*np.exp((-0.5)*pow(((x-miu)/std),2))
```

```python
# Generacion de arreglo de valores entre -4 y 4 con un paso de 0.1
x=np.arange(-4,4,0.1)
y=my_gaussian(x,2,1)
sns.lineplot(x=x,y=y)
plt.show()
```

```python
#Imputs dr todas las funciones
dist = norm(1,1) # promedio 0 y std 1
x=np.arange(-4,4,0.1)
# Calculo de propabilidades
y= [dist.pdf(i) for i in x]
sns.lineplot(x=x,y=y)
```

```python
dist = norm(0,1) # promedio 0 y std 1
x=np.arange(-4,4,0.1)
y= [dist.cdf(i) for i in x]
sns.lineplot(x=x,y=y)
```

```python
# Data Frame - Excel
# df.head(5) - Show first five rows
# Creamos un dataset reducido a la columna que vamos a procesar
df = pd.read_excel('normal_real_example.xls')
wing = df[['Normally Distributed Housefly Wing Lengths']].copy()
# Con iloc - limpiamos los datos de la columnnas que no son numericos , Que aparecen al descomentar df.head(5)
wing_clean=wing.iloc[3:]
values, frecuencia = np.unique(wing_clean.values, return_counts=True)
# wing_clean es un objeto numpy
miu=wing_clean.values.mean() # promedio
std=wing_clean.values.std() # desviacion standar

# Promedio y desviacion estandar calculada de los datos miu - std
dist_norm = norm(miu,std)
# Calcula de densidad de probabilidad para los valores generados en x
x_generate=np.arange(30,60,0.1)
y_teorica=[dist_norm.pdf(value) for value in x_generate]

plt.bar(values,(frecuencia/len(wing_clean)))
plt.plot(x_generate,y_teorica)
plt.show()

# Con seaborn | No logre temporalmente sobreponer los graficos
'''
sns.barplot(x=values, y=frecuencia/len(wing_clean), errorbar=None)
sns.lineplot(x=x_generate, y=y_teorica)
plt.show()
'''
```
