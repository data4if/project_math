```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
```

```python
def gaussian(x,miu,std):
    return (1/(std*((2*np.pi)**(1/2))))*np.exp(-0.5*((x-miu)/std)**(1/2))
```

```python
def gaussion_platzi(x,miu,std):
    return (1/(std*np.sqrt(2*np.pi)))*np.exp((-0.5)*pow(((x-miu)/std),2))
```

```python
x=np.arange(-4,4,0.1)
#y=gaussian(x,0.0,1.0)
y=gaussion_platzi(x,0,1)
sns.lineplot(x=x,y=y)
```

```python
dist = norm(2,1) # promedio 0 y std 1
x=np.arange(-4,4,0.1)
y= [dist.pdf(i) for i in x]
sns.lineplot(x=x,y=y)
```

```python
dist = norm(0,1) # promedio 0 y std 1
x=np.arange(-4,4,0.1)
y= [dist.cdf(i) for i in x]
sns.lineplot(x=x,y=y)
```
