```python
import timeit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model


# Tanto X como y, Son tuplas de python
X, y = datasets.load_diabetes(return_X_y=True)
# Preguntar a chat gpt 3 como funciona este algoritmo
# X.shape()
raw = X[:,None,2]
# raw.shape()
```

```python
print(y[0])
print(X[0])
print(raw[0])
```

```python
# Reglas de escalamiento 
max_raw=max(raw)
min_raw=min(raw)
scaled =(2*raw -max_raw - min_raw) / (max_raw - min_raw)
sca_z=(raw-np.median(raw))/(np.std(raw))
```

```python
fig, axs = plt.subplots(2,1, sharex=True)
axs[0].hist(raw)
axs[1].hist(scaled)
plt.show()
```

```python
def train_raw():
    linear_model.LinearRegression().fit(raw,y)
    
def train_scaled():
    linear_model.LinearRegression().fit(scaled,y)
                                        
def train_scaled_z():
    linear_model.LinearRegression().fit(sca_z,y)
```

```python
raw_time=timeit.timeit(train_raw, number=100)
scaled_time=timeit.timeit(train_scaled, number=100)
scaled_time_z=timeit.timeit(train_scaled_z, number=100)

print(raw_time)
print(scaled_time)
print(scaled_time_z)
```
