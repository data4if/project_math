## MLE como base para la regrsion logistica

- estamos trabajando con la distribucion de bernoulli

```python
def likelihood(y,yp):
    return yp*y+(1-y)*(1-yp)
```

```python
y=np.arange(0,1,0.01)
yp=np.arange(0,1,0.01)
Y,YP =np.meshgrid(y,yp) #Construye una cuadricula
Z = likelihood(Y,YP)
```

```python
fig = go.Figure(data=[go.Surface(z=Z, x=Y, y=YP)])
fig.show()
```
